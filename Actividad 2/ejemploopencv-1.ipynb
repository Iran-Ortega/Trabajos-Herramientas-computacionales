Este es mi proyecto, es un ejemplo de como utilizar OpenCV de manera efectiva. 

```python
#importar librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt


```

```python
bgr = cv2.imread('porsche-911.jpg.webp')
print("El tamaño de la imagen es: ", bgr.shape)

#Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

#Mostrar la imagen
plt.imshow(rgb)
```

```python
#Mostrar la imagen en escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#Mostrar los componentes de color de la imagen
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Canal Azul')
plt.subplot(1,3,2)
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Canal Verde')
plt.subplot(1,3,3)
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Canal Rojo')
```

```python
#Mostrar los componentes HLS de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(hls[:,:,0], cmap='gray')
plt.title('Canal H')
plt.subplot(1,3,2)
plt.imshow(hls[:,:,1], cmap='gray')
plt.title('Canal L')
plt.subplot(1,3,3)
plt.imshow(hls[:,:,2], cmap='gray')
plt.title('Canal S')
```
